## Operaciones en Pandas

Vamos creando una columan nueva de ganancias que será igual a precio menos costo de precios y costos

```python
df_precio['Precio'] = df_precio['Precio'].str.replace('$','').astype(float)
```

```python
df_precio.head()
```

```python
df_precio['Costo'] = df_precio['Costo'].str.replace('$','').astype(float)
```

```python
df_precio.head(5)
```

```python
df_precio['Ganacia'] = df_precio['Precio'] - df_precio['Costo']
df_precio.head()
```

Vamos tirando la columna de "hora" del restaruant

```python
df_restauran = df_restauran.drop('Hora de Cobro', axis=1)
```

```python
df_restauran.head(6)
```

Arma un nuevo dataframe con todos los platillos de más de 100 pesos de precio (usando filtros)

```python
df_precios_caros = df_precio[df_precio['Precio'] > 100]
```

```python
df_precios_caros.head(20)
```

Arma una nueva columna que clasifique platillos en alto/bajo margen, según si su ganancia es mayor a 100 pesos

```python
df_precio['Marge'] = np.where(df_precio['Ganacia'] > 100, 'Alto','Bajo')
```

```python
df_precio.head(20)
```

Cambiale el nombre a la columna de Atendió por Mesero en el df de Restaurant

```python
df_restauran = df_restauran.rename(columns={'Atendió':'Mesero'})
```

```python
df_restauran.head(10)
```

## Merging o Vlookup

Vamos a juntar los 2 dataframes usando un left join

```python
df_rest_precio = df_restauran.merge(df_precio, on=['Producto'], how='left')
```

## Tablas Dinámicas

Pandas también nos permite analizar datos mediante tablas dinámicas

Vamos creando una tabla dinámica que nos muestre las ventas por categoría

```python
venxcat = pd.pivot_table(df_rest_precio,values='Precio',index='Categoria', aggfunc='sum')
```

```python
venxcat
```

Ahora una tabla dinámica que nos muestre las ventas por categoría y por tipo en la columna

```python
venxcatxtipo = pd.pivot_table(df_rest_precio, values='Precio',index='Categoria', columns='Tipo_x', aggfunc='sum')
```

```python
venxcatxtipo
```

Crea una tabla dináica que muestre las ventas por tipo y categoría (filas), y por mesero (columnas)

```python
venxtipocatxmesero = pd.pivot_table(df_rest_precio, values='Precio',index=['Tipo_x','Categoria'], columns='Mesero', aggfunc='sum')
```

```python
venxtipocatxmesero
```

Crea una tabla dinámica que muestre las ventas promedio por mesero

```python
venxpromxmesero = pd.pivot_table(df_rest_precio, values='Precio', index='Mesero')
```
